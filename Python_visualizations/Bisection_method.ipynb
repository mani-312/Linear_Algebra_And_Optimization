```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython import display
import time
```

```python

# f(x) = -5x^5 + 4x^4 - 12x^3 + 11x^2 - 2x + 1
# Visualize the function at https://www.desmos.com/calculator
f = np.poly1d([-5,4,-12,11,-2,1])

# Derivative of f(x)
d = np.polyder(f)

# Interval of uncertainity [a,b]
a = -0.5
b = 0.5

# K = Number of iteration till convergence
k = 0


while (b-a)>1e-5:
    print(a,b)
    k = k+1
    mid = (a+b)/2
    der = d(mid)
    if der == 0:
        break
    # Minimum is lying to right of mid
    elif der<0:
        a = mid

    # Minimum is lying to left of mid
    else:
        b = mid
x = (a+b)/2
print("\nMinimum point ",x)
print("Minimum value ",f(x))
print("Number of iterations ",k)
```

```python
f = np.poly1d([2,-10,15])
d = np.polyder(f)

a = -6
b = 6

k = 0
ar = np.arange(-7,7,0.1)


while k<=8:

    sns.relplot(x = ar,y = f(ar),height = 7)
    plt.plot([a,a], [0,f(a)], linewidth=3,color = 'green')
    plt.plot([b,b], [0,f(b)], linewidth=3,color = 'green')
    
    display.display(plt.gcf())
    display.clear_output(wait=True)
    time.sleep(2.0)
    
    
    k = k+1
    mid = (a+b)/2
    der = d(mid)

    if der == 0:
        break
    elif der<0:
        a = mid
    else:
        b = mid
```
