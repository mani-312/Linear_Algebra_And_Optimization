```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython import display
import time
import math
```

```python
# f(x) = (1/4)x^4 − (5/3)x^3 − 6x^2 + 19x − 7
f = np.poly1d([1/4,-5/3,-6,19,-7])
g = np.polyder(f)
g1 = np.polyder(g)
xk = -3
k = 0

while abs(g(xk)) > 1e-5 and k<100:
    print(xk,g(xk))
    prev = f(xk)
    k = k+1
    xk = xk - (g(xk)/g1(xk))
print(xk,g(xk))
```

```python
ar = np.arange(-4,4,0.1)
sns.set_style("darkgrid")


xk = -1.5
k = 0

while abs(g(xk)) > 1e-5 and k<=5:
    
    k = k+1
    xk_new = xk - (g(xk)/g1(xk))
    
    sns.relplot(x = ar,y = g(ar),height = 7) # g(x) plot
    plt.plot([-4,4],[0,0],color = "green",linewidth = 3) # x-axis line
    plt.plot([xk,xk_new],[g(xk),0],color = "pink",linewidth = 3) # tangent at xk
    plt.plot([xk_new,xk_new],[0,g(xk_new)],color = "yellow",linewidth = 3)
    plt.text(xk, 0.5, "x"+str(k),fontsize = 13)
    plt.text(xk_new, 0.5, "x"+str(k+1),fontsize = 13)
    
    display.display(plt.gcf())
    display.clear_output(wait=True)
    time.sleep(2.0)
    
    xk = xk_new
```

```python

# f(x) = log(e^x+e^(-x))
def g(x):
    return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))

def g1(x):
    ret = 1- ((math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x)))**2
    if ret == 0:
        ret = 0.01
    return ret

ar = np.arange(-10,10,0.1)
sns.set_style("darkgrid")


xk = 1.2
k = 0

while abs(g(xk)) > 1e-5 and k<=2:
    
    k = k+1
    xk_new = xk - (g(xk)/g1(xk))
    
    
    sns.relplot(x = ar,y = [g(i) for i in ar],height = 7) 
    plt.plot([-10,10],[0,0],color = "green",linewidth = 3)
    plt.plot([xk,xk_new],[g(xk),0],color = "pink",linewidth = 3)
    plt.plot([xk_new,xk_new],[0,g(xk_new)],color = "yellow",linewidth = 3)
    plt.text(xk, 0.05, "x"+str(k),fontsize = 13)
    plt.text(xk_new, 0.05, "x"+str(k+1),fontsize = 13)
    
    display.display(plt.gcf())
    display.clear_output(wait=True)
    time.sleep(2.0)
    
    xk = xk_new
```
